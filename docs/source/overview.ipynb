# Overview

```python
import ultraplot as uplt
from gmrtbeam.core import GMRTBeam
from gmrtbeam.fit import BeamFitter
```

```python
beam = GMRTBeam.new(
    f0=650.0,
    ra="04h30m49.1s",
    dec="+64d50m53.0s",
    date="14/01/25",
    time="15:28:16",
    mode="PC",
    gac=[
        "C00",
        "C01",
        "C02",
        "C03",
        "C04",
        "C05",
        "C06",
        "C08",
        "C09",
        "C10",
        "C11",
        "C12",
        "C13",
        "C14",
        "E02",
        "E03",
        "S01",
        "S02",
        "S03",
        "W01",
        "W02",
        "W03",
    ],
)
```

```python
beam.compute()
```

```python
beam.plot()
```

```python
gfitter = BeamFitter(beam, model="gaussian")
efitter = BeamFitter(beam, model="elliptical")
```

```python
gfitter.fit()
efitter.fit()
```

```python
fig = uplt.figure(width=7.5, height=7.5)
ax = fig.subplot()
beam.plot(ax=ax)
gfitter.plot(ax=ax, plotbeam=False)
efitter.plot(ax=ax, plotbeam=False)
ax.legend(loc="ur")
uplt.show()
```
