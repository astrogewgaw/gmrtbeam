# Overview

```python
from gmrtbeam.fit import BeamFitter
from gmrtbeam.mosaic import GMRTMosaic
from gmrtbeam.core import BeamMode, GMRTBeam
```

```python
beam = GMRTBeam(
    f0=650.0,
    rastr="04h30m49.1s",
    decstr="+64d50m53.0s",
    datestr="14/01/25",
    timestr="15:28:16",
    mode=BeamMode.PC,
    gac=[
        "C00",
        "C01",
        "C02",
        "C03",
        "C04",
        "C05",
        "C06",
        "C08",
        "C09",
        "C10",
        "C11",
        "C12",
        "C13",
        "C14",
        "E02",
        "E03",
        "S01",
        "S02",
        "S03",
        "W01",
        "W02",
        "W03",
    ],
)
```

```python
beam.compute()
```

```python
beam.plot()
```

```python
beam.plotuv()
```

```python
beam.plotsky()
```

```python
beam.to_fits("/home/upanda/Downloads/beam.fits")
```

```python
gfitter = BeamFitter(beam, model="gaussian")
efitter = BeamFitter(beam, model="elliptical")
```

```python
gfitter.fit()
efitter.fit()
```

```python
gfitter.plot()
```

```python
efitter.plot()
```

```python
mosaic = GMRTMosaic(nbeams=160, beam=beam, fitter=efitter)
mosaic.tile(recenter=False)
```

```python
mosaic.plot()
```

```python
mosaic.to_fits("/home/upanda/Downloads/tile.fits")
```
